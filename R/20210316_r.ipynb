## Q1: 자동차 배기량에 따라 고속도로 연비가 드는지 알아보려고 한다. displ(배기량)이 4이하인 자동차와 5이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비)가 평균적으로 더 높은지 알아봐라 

```r
mpg_a <- mpg %>% filter(displ<=4)
mpg_b <- mpg %>% filter(displ>=5)
```

```r
mean(mpg_a$hwy)
```

```r
mean(mpg_b$hwy)
```

## 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다.'audi'와 'toyota' 중 어느 manufacurer(자동차 제조회사)의 cty(도시 연비)가 평균적으로 더 높은지 알아보세요.

```r
mpg_audi <- mpg %>% filter(manufacturer=='audi')
mpg_toyota <- mpg %>% filter(manufacturer=='toyota')
```

```r
mean(mpg_audi$cty)
```
